```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
### Create a list of Column Names (the monthly dates) we can slap on each import
column_name_list = pd.read_excel("SP500_Index_Data_values_only.xlsx", sheet_name="Ticker", 
                   header=0, usecols="R:JD", skiprows=2, nrows=1, 
                   na_values=None, na_filter=False, parse_dates=True).columns.to_list()

print(column_name_list[:5])
print(column_name_list[-5:])
```

```python
### Create a ticker_df that will be the starting point of our loop to merge the other sheets of the excel file
ticker_df = pd.read_excel("SP500_Index_Data_values_only.xlsx", sheet_name="Ticker", 
                   names=column_name_list, header=0, usecols="R:JD", skiprows=10, nrows=505, 
                   na_values=None, keep_default_na=False, na_filter=False, parse_dates=True)

ticker_df
```

```python
### Put our starting point in "long format" with melt
### Loop through the data items we want in the excel workbook and concat them

excel_data_df = ticker_df.melt( var_name='Date', value_name='Ticker')

excel_sheet_list = [
                    'Name',
                    'Sector',
                    'Price',
                    'Local_Returns_12m',
                    'Local_Returns_1m',
    
                    'Trail_DivYld',
                    'PB',
                    'Trail_EV_EBITDA',
                    'Trail_PE',
                    'Trail_EPS',
                    'Trail3yrAvg_EPSgro',
                    'Trail3yrAvg_DPSgro',
                    'Volatility',
                    'Debt_to_MktCap',
                    'NetDebt_EBITDA',
                                       ]

for sheet in excel_sheet_list:
    print(sheet)
    temp_df = pd.read_excel("SP500_Index_Data_values_only.xlsx", sheet_name=sheet, 
                   names=column_name_list, header=0, usecols="R:JD", skiprows=10, nrows=505, 
                   na_values=None, keep_default_na=False, na_filter=False, parse_dates=True)
    
    temp_df = temp_df.melt( var_name='Date', value_name=sheet).drop(['Date'], axis=1)
    
    excel_data_df = pd.concat([excel_data_df, temp_df], axis=1)
    
del temp_df
```

```python
### .info() for sanity check
excel_data_df.info()
```

```python
### Peek at the dataframe for sanity check
excel_data_df
```

Basic Data Cleaning

```python
### We know there are several "junk" values in the excel workbook that we need to remove...
### replace those values with np.nan
### remove any 'Ticker' that doesn't exist (had "Junk" in the cell)

df = excel_data_df.copy()
df.replace(to_replace=['x', '#N/A N/A', 
                       '#N/A Invalid Security', '', 
                       '#N/A Mandatory parameter [SECURITY] cannot be empty', 
                       'na',
                       '#N/A Requesting Data...',
                       ],
           value=np.nan, inplace=True, limit=None, regex=False, method='pad')

df = df[~df['Ticker'].isna()]

#df
```

```python
### We know for certain columns we want to 'force' a float data type
for column in df.columns[4:]:
    df[column] = df[column].astype(float)
```

```python
### Sanity check for the columns we wante to be float data type
df.info()
```

```python
### Our Ticker column has some extraneous or unnecessary values we can get rid of:
### We use a regex to do so:
df['Ticker'] = df.Ticker.str.replace(' U. Equity', '', n=- 1, case=None, flags=0, regex=True)
#df
```
